```python
import pandas as pd

data = pd.read_csv('pizza_sales.csv')

# Convert date
data['order_date'] = pd.to_datetime(data['order_date'], dayfirst=True, errors='coerce')
```

```python
data['MonthYear'] = data['order_date'].dt.strftime("%B %Y")  # "December 20215"
data['MonthNum'] = data['order_date'].dt.month    # "Extracts month as number (1-12) e,g 2015-12-10 - 12"
data['WeekdayName'] = data['order_date'].dt.day_name   # Gets full day name 2015-12-31 â†’ "Thursday"
data['WeekdayNum'] = data['order_date'].dt.weekday
data['OrderHour'] = data['order_date'].dt.hour
```

```python
print(data.columns)
```

## 1. Top 5 Best-Selling Pizzas by Quantity & Revenue

```python
import os

print("="*60)
print("1: TOP 5 BEST-SELLING PIZZAS")
print("="*60)

# By Quantity
top5_quantity = data.groupby('pizza_name').agg({
    'quantity': 'sum',
    'total_price': 'sum',
    'unit_price': 'mean'
    }).sort_values('quantity', ascending=False).head(5)

print("Top 5 Pizzas by Quantity")
print(top5_quantity[['quantity', 'total_price']])


# By Revenue
top5_revenue = data.groupby('pizza_name').agg({
    'quantity': 'sum',
    'total_price': 'sum',
    'unit_price': 'mean'
    }).sort_values('total_price', ascending=False).head(5)

print("\nTop 5 Pizzas by Revenue")
print(top5_revenue[['quantity', 'total_price']])

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Top by Quantity
top5_quantity['quantity'].plot(kind='bar', ax=ax1, color='#d52d2d')
ax1.set_title('Top 5 Pizzas by Quantity Sold', fontsize=14, fontweight='bold')
ax1.set_ylabel('Number of Pizzas Sold')
ax1.tick_params(axis='x', rotation=45)

# Top by Revenue
top5_revenue['total_price'].plot(kind='bar', ax=ax2, color='#d67819')
ax2.set_title('Top 5 Pizzas by Revenue Generated', fontsize=14, fontweight='bold')
ax2.set_ylabel('Revenue ($)')
ax2.tick_params(axis='x', rotation=45)

# Adjust layout and save
plt.tight_layout()
# Create the directory if it doesn't exist
if not os.path.exists('outputs'):
    os.makedirs('outputs')
plt.savefig("outputs/best_selling.png")
plt.show()
```

##2: Pizza Categories Contribution

```python
print("="*60)
print("2: PIZZA CATEGORIES PERFORMANCE")
print("="*60)

# By Category
category_performance = data.groupby('pizza_category').agg({
    'total_price': 'sum',
    'quantity': 'sum',
    'order_id': 'nunique'
    }).sort_values('total_price', ascending=False).head(5)

print("Category Performance")
print(category_performance)

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Revenue Share Pie Chart
colors = ['#dc842b', '#473021', '#bbafa7', '#e8aa84', '#2c1608']
ax1.pie(category_performance['total_price'], labels=category_performance.index,
        autopct='%1.1f%%', startangle=90, colors=colors)
ax1.set_title('Sales by Pizza Category', fontsize=14, fontweight='bold')

# Quantity Comparison
category_performance['quantity'].plot(kind='bar', ax=ax2, color='orange')
ax2.set_title('Pizzas Sold by Category', fontsize=14, fontweight='bold')
ax2.set_ylabel('Number of Pizzas Sold')
ax2.tick_params(axis='x', rotation=45)

# Adjust layout and save
plt.tight_layout()
plt.savefig("outputs/category_performance.png")
plt.show()
```

## 3: Pizza Sizes Demand

```python
print("="*60)
print("3: SALES BY PIZZA SIZES")
print("="*60)

size_demand = data.groupby('pizza_size').agg({
    'total_price': 'sum',
    'quantity': 'sum',
    'order_id': 'nunique'
    }).sort_values('quantity', ascending=False)

print("Sales By Pizza Size")
print(size_demand)


# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

# Sales by Pizza Size Pie Chart
colors = ['#dc842b', '#473021', '#bbafa7', '#e8aa84', '#2c1608']
ax1.pie(size_demand['total_price'], labels=size_demand.index,
        autopct='%1.1f%%', startangle=90, colors=colors)
ax1.set_title('Sales by Pizza Size', fontsize=14, fontweight='bold')

# Size Demand Comparison
size_demand['quantity'].plot(kind='bar', ax=ax2, color='orange')
ax2.set_title('Pizzas Sold by Size', fontsize=14, fontweight='bold')
ax2.set_ylabel('Size')
ax2.tick_params(axis='x', rotation=45)

# Adjust layout and save
plt.tight_layout()
plt.savefig("outputs/sales_by_sizes.png")
plt.show()
```

## 4. The bottom 5 worst-selling pizzas

```python
print("="*60)
print("4: BOTTOM 5 WORST-SELLING PIZZAS")
print("="*60)

# By Quantity
bottom5_quantity = data.groupby('pizza_name').agg({
    'quantity': 'sum',
    'total_price': 'sum',
    'unit_price': 'mean'
    }).sort_values('quantity', ascending=True).head(5)

print("Bottom 5 Pizzas by Quantity")
print(bottom5_quantity[['quantity', 'total_price']])


# By Revenue
bottom5_revenue = data.groupby('pizza_name').agg({
    'quantity': 'sum',
    'total_price': 'sum',
    'unit_price': 'mean'
    }).sort_values('total_price', ascending=True).head(5)

print("\nBottom 5 Pizzas by Revenue")
print(bottom5_revenue[['quantity', 'total_price']])

# Visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Bottom by Quantity
bottom5_quantity['quantity'].plot(kind='bar', ax=ax1, color='#d52d2d')
ax1.set_title('Bottom 5 Pizzas by Quantity Sold', fontsize=14, fontweight='bold')
ax1.set_ylabel('Number of Pizzas Sold')
ax1.tick_params(axis='x', rotation=45)

# Bottom by Revenue
bottom5_revenue['total_price'].plot(kind='bar', ax=ax2, color='#d67819')
ax2.set_title('Bottom 5 Pizzas by Revenue Generated', fontsize=14, fontweight='bold')
ax2.set_ylabel('Revenue ($)')
ax2.tick_params(axis='x', rotation=45)

# Adjust layout and save
plt.tight_layout()
plt.savefig("outputs/worst_selling.png")
plt.show()
```

## 5. Is there a correlation between price and popularity?

```python
grouped = data.groupby('pizza_name').agg({
    'quantity': 'sum',        # Total pizzas sold (popularity)
    'unit_price': 'mean'      # Average price per pizza
}).reset_index().sort_values(by='quantity', ascending=False);

print(grouped.head())
```

```python
correlation = grouped['unit_price'].corr(grouped['quantity'])
print(f"Correlation between price and popularity: {correlation:.2f}")
```

Thereâ€™s a moderate negative relationship between price and popularity.
As the price goes up, sales quantity tends to go down.


# 5. Sales Trend

## 5.1 Sales trends on a monthly basis

```python
monthly_sales = data.groupby(data['order_date'].dt.to_period('M')).agg({
    'order_id': 'nunique',
    'total_price': 'sum'
}).reset_index()

monthly_sales.columns = ['Month', 'Orders', 'Revenue']

print(monthly_sales.head())

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Month'].astype(str), monthly_sales['Orders'], marker='o', linestyle='-', color='#d67819', label='Orders')
plt.title('Monthly Orders Trend', fontsize=14, fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Save before showing
plt.savefig("outputs/monthly_orders_trend.png")
plt.show()

best_month = monthly_sales.loc[monthly_sales['Orders'].idxmax()]
worst_month = monthly_sales.loc[monthly_sales['Orders'].idxmin()]

print(f"Best Month: {best_month['Month']} with {best_month['Orders']} orders")
print(f"Worst Month: {worst_month['Month']} with {worst_month['Orders']} orders")

```

## 5.2 Day of the week with the highest number of orders

```python
weekly_sales = data.groupby(data['order_date'].dt.day_name()).agg({
    'order_id': 'nunique',
    'total_price': 'sum'
}).reset_index()

weekly_sales.columns = ['Weekday', 'Orders', 'Revenue']

print(weekly_sales)

highest_day_of_the_week = weekly_sales.loc[weekly_sales['Orders'].idxmax()]
lowest_day_of_the_week = weekly_sales.loc[weekly_sales['Orders'].idxmin()]

print(f"\nDay of the week with the highest number of orders: {highest_day_of_the_week['Weekday']} with {highest_day_of_the_week['Orders']} orders")
print(f"\nDay of the week with the lowest number of orders: {lowest_day_of_the_week['Weekday']} with {lowest_day_of_the_week['Orders']} orders")


#Visualization
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
orders_per_day = weekly_sales.set_index('Weekday').reindex(weekday_order)['Orders']

plt.figure(figsize=(10, 6))
orders_per_day.plot(kind='bar', color='#d67819')

plt.title('Orders by Day of the Week', fontsize=14, fontweight='bold')
plt.xlabel('Day of the Week')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()

# Save before showing
plt.savefig("outputs/orders_by_day_of_the_week.png")
plt.show()
```

## 5.3 The busiest hour of the day

```python
data['order_time'] = pd.to_datetime(data['order_time'], format='%H:%M:%S')
data['hour'] = data['order_time'].dt.hour

orders_per_hour = data.groupby('hour')['order_id'].nunique().sort_values(ascending=False)
print(orders_per_hour)

#Visualization

plt.figure(figsize=(10, 6))
orders_per_hour.plot(kind='bar', color='#d67819')

plt.title('Orders by hour of the Day', fontsize=14, fontweight='bold')
plt.xlabel('Hour of the Day')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()

# Save before showing
plt.savefig("outputs/orders_by_hour_of_the_day.png")
plt.show()


busiest_hour = orders_per_hour.idxmax()
busiest_value = orders_per_hour.max()
print(f"Busiest hour of the day: {busiest_hour}:00 with {busiest_value} orders")
```
